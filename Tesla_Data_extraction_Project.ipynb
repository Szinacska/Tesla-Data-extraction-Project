```python
tesla_stock = yf.Ticker("TSLA")
tesla_stock
```

```python
tesla_data = tesla_stock.history(period="max")
```

```python
tesla_data.reset_index(inplace=True)
tesla_data.head(5)
```

```python
tesla_financials = tesla_stock.financials
tesla_revenue = tesla_financials.loc['Total Revenue']
print(tesla_revenue.tail())
```

```python

gme_stock = yf.Ticker("GME")
gme_data = gme_stock.history(period="max")
gme_data_reset = gme_data.reset_index(drop=True)
gme_data_reset.to_csv('gme_data_reset.csv', index=False)
print(gme_data_reset.head())
```

```python
import plotly.graph_objects as go
def make_graph(data):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Tesla Closing Price'))
    fig.update_layout(
        title='Tesla Stock Price Over the Last Year',
        xaxis_title='Date',
        yaxis_title='Price (USD)',
        template='plotly_dark'
    )
    fig.show()
make_graph(tesla_data)
```
